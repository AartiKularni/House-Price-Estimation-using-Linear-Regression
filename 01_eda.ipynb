```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
import numpy as np


```

```python
df = pd.read_csv("data/train.csv",
                 usecols=["SalePrice","GrLivArea","BedroomAbvGr","FullBath","HalfBath"])
```

```python
display(df.info())          # <–– note data types & NaN counts
display(df.describe().T)    # <–– look at mean vs. median, skew
```

```python
sns.histplot(df["SalePrice"], kde=True)
plt.title("SalePrice – raw distribution"); plt.show()
```

```python
sns.pairplot(df, height=2.5)
plt.suptitle("Scatter-matrix of raw features", y=1.02)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues", vmin=-1, vmax=1)
plt.title("Correlation matrix"); plt.show()
```
